# Processing of VASP NEB data to produce Figs 4, 5, 6 & 7 in the main manuscript
    
This notebook reads in data from from the `.csv` files in `open_data/data/for_plotting` and generates figures 4, 5, 6, and 7 in the manuscript &ldquo;Interfacial Strain Effects on Lithium Diffusion Pathways in the Spinel Solid Electrolyte Li-Doped MgAl<sub>2</sub>O<sub>4</sub>&rdquo;
  
    
## Required files:
```
open_data/
    ├── data/
    |   ├── for_plotting/
    |       ├── aniso_0.25.csv
    |       ├── aniso_0.50.csv
    |       ├── spinel_aniso_data.csv
    |       ├── spinel_iso_data.csv
    |       └── spinel_iso_pot_vol.csv
    ├── analysis/
        ├── pathway_figs/
            ├── 1_0.25.png
            ├── 1_0.5.png
            ├── 2_0.25.png
            ├── 2_0.5.png
            ├── 3R_0.25.png
            ├── 3R_0.5.png
            ├── 3_0.25.png
            └── 3_0.5.png
```

## Outputs:
- `NEB_ISO.pdf`
- `aniso_neb_PATH.pdf`
- `aniso_neb_0.5.pdf`
- `aniso_neb_0.25.pdf`
    

````python
import pandas as pd
import numpy as np

import matplotlib.ticker as ticker
from scipy.optimize import curve_fit
from matplotlib.ticker import FormatStrFormatter
import matplotlib.pyplot as plt
# import matplotlib
from matplotlib import rcParams
import figure_formatting
from figure_formatting import get_y_limits, get_y_axis_limits, interpolate_df, place_image, line, set_rcParams
from figure_formatting import colours, symbol, nearly_black, master_formatting

%matplotlib inline
%config InlineBackend.figure_format='retina'

set_rcParams( master_formatting )
````

````python
def pathway_label( strain, conc, pathway, aniso=False ):
    if aniso:
        label = r"$^\mathrm{{{}}}${}$_{{{:4.2f}}}^\mathrm{{{}}}$".format( 'A', strain, conc, pathway )
    else:
        label = r"{}$_{{{:4.2f}}}^\mathrm{{{}}}$".format( strain, conc, pathway )
    return label
````

````python
pathway_dir = 'pathway_figs/pngs'
figures_dir = '../figures'
data_dir = '../data/for_plotting'

data_spinel = pd.read_csv( '{}/spinel_iso_data.csv'.format( data_dir ), delim_whitespace = True,comment='#')
````

````python
'''
   Produces Fig 4 in manuscript
'''

neb_params = { 'font.size': 10,
               'axes.labelsize': 16,
               'ytick.labelsize': 14 }
set_rcParams( neb_params )

ncols=4
nrows=4

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize = (12, 10))

for i, path in enumerate( [ '1','2','3','3R'] ):
    for j,conc in enumerate ([0.25,0.50]):
                
        e_axis = axes[(j*2),i]
        p_axis = axes[(j*2)+1,i]

        if i==0:
            e_axis.set_ylabel(r'$\Delta E_\mathrm{NEB}$ [eV]')
            p_axis.set_ylabel(r'$-\Delta\Phi_\mathrm{Li}$ [V]')
        if i>0:
            e_axis.set_yticklabels([])
            e_axis.tick_params( left='off' )
            p_axis.set_yticklabels([])
            p_axis.tick_params( left='off' )

        e_axis.set_xlim([0,10.0])        
        e_axis.set_ylim( get_y_limits(1.2,0.5) )
        e_axis.set_yticks(np.arange(-0.6,+0.61,0.2))
        e_axis.set_xticks(np.arange(0,11,1))
        e_axis.set_xticklabels([])
        e_axis.tick_params( bottom='off' )
        
        p_axis.set_xlim([0,10.0]) 
        p_axis.set_ylim( get_y_axis_limits(data_spinel.loc[ (data_spinel['conc']==conc)],'d_pot',0.5,1,0.15) )
        p_axis.set_xticks(np.arange(0,11,1))
        p_axis.set_xticklabels([])
        p_axis.tick_params( bottom='off' )

        for l, strain in enumerate(['U','LM','LT']):
            label=pathway_label( strain, conc, path )            
            x,y = interpolate_df(data_spinel.loc[(data_spinel['path'] == path) 
                        & (data_spinel['conc'] == conc) 
                        & (data_spinel['strain'] == strain)]['delta_E'],[0,10],0.01)
            #print(data_spinel.loc[(data_spinel['path'] == path) 
            #            & (data_spinel['conc'] == conc) 
            #            & (data_spinel['strain'] == strain)]['delta_E'])
            e_axis.plot(x, y, c=colours[strain], label=label )
            e_axis.legend( loc='lower left' )           
            
            x,y = interpolate_df(data_spinel.loc[(data_spinel['path'] == path) 
                        & (data_spinel['conc'] == conc) 
                        & (data_spinel['strain'] == strain)]['d_pot'],[0,10],0.01)
            p_axis.plot(x,y,c=colours[strain],linestyle='dashed', label=label )
            p_axis.legend( loc='lower left' )

fig.tight_layout()

# Add pathway schematic png figures
xyz = ( 0.265, 0.45, 0.04 )
for i, e in enumerate( [ '1', '2', '3','3R'] ):
    for j, conc in enumerate (['0.25','0.50']):
        pot_index=(j*2)+1
        ene_index=(j*2)
        image_filename="{}/{}_{}.png".format(pathway_dir, e, conc)
        place_image(fig, axes[ene_index,i], image_filename, *xyz )
        place_image(fig, axes[pot_index,i], image_filename, *xyz )  
fig.savefig( filename = '{}/neb_pathways_iso.pdf'.format( figures_dir ) )  

set_rcParams( master_formatting )
````

````python
'''
   Produces Fig 7 in manuscript
'''
set_rcParams( neb_params )

data_aniso_spinel = pd.read_csv( '{}/spinel_aniso_data.csv'.format( data_dir ), delim_whitespace=True, comment='#')

ncols = 4
nrows = 4

conc = 0.25

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize = (12, 10))

for i, path in enumerate( [ '1','2','3','3R'] ):
    for j,strain in enumerate (['LM','LT']):
        
        e_axis = axes[(j*2),i]
        p_axis = axes[(j*2)+1,i]
        
        for s in [ 'top', 'bottom', 'left', 'right' ]:
            e_axis.spines[s].set_visible( False )
            p_axis.spines[s].set_visible( False )

        if i==0:
            e_axis.set_ylabel(r'$\Delta E_\mathrm{NEB}$ [eV]')
            p_axis.set_ylabel(r'$-\Delta\Phi_\mathrm{Li}$ [V]') 
        if i>0:
            e_axis.set_yticklabels([])
            e_axis.tick_params( left='off' )
            p_axis.set_yticklabels([])
            p_axis.tick_params( left='off' )
                 
        e_axis.set_xlim([0,10.0])        
        e_axis.set_ylim(get_y_limits(1.0,0.4))
        e_axis.set_xticks(np.arange(0,11,1))
        e_axis.set_xticklabels([])
        e_axis.tick_params( bottom='off' )

        p_axis.set_xlim([0,10.0]) 
        p_axis.set_ylim(get_y_axis_limits(data_spinel.loc[ (data_spinel['conc']==conc)],'d_pot',0.4,1,0.15))
        p_axis.set_xticks(np.arange(0,11,1))
        p_axis.set_xticklabels([])
        p_axis.tick_params( bottom='off' )
            
        for strain_axis in range(1,4):
                
            x,y = interpolate_df(data_aniso_spinel.loc[(data_aniso_spinel['path'] == path) 
                            & (data_aniso_spinel['conc'] == conc) 
                            & (data_aniso_spinel['strain'] == strain)
                            & (data_aniso_spinel['axes'] == strain_axis)]['delta_E'],[0,10],0.01)
            if strain_axis == 1:
                label = pathway_label( strain, conc, path, aniso=True )
            else:
                label = None
                     
            e_axis.plot(x,y,c=colours[strain],label=label)
            
            x,y = interpolate_df(data_aniso_spinel.loc[(data_aniso_spinel['path'] == path) 
                        & (data_aniso_spinel['conc'] == conc) 
                        & (data_aniso_spinel['strain'] == strain)
                        & (data_aniso_spinel['axes'] == strain_axis)]['d_pot'], [0,10], 0.01 )
            p_axis.plot(x,y,c=colours[strain],linestyle='dashed', label=label)
            
        col=colours['U']     
        x,y = interpolate_df(data_spinel.loc[(data_spinel['path'] == path) 
                        & (data_spinel['conc'] == conc) 
                        & (data_spinel['strain'] == 'U')]['delta_E'],[0,10],0.01)
        e_axis.plot( x, y, c=col, label=pathway_label( 'U', conc, path ) )
        e_axis.legend(loc='lower left')
        
        x,y = interpolate_df(data_spinel.loc[(data_spinel['path'] == path) 
                        & (data_spinel['conc'] == conc) 
                        & (data_spinel['strain'] == 'U')]['d_pot'], [0,10], 0.01 )
        p_axis.plot(x,y,c=col,linestyle='dashed', label=pathway_label( 'U', conc, path ) )         
        p_axis.legend(loc='lower left')
                        
fig.tight_layout()        

# Add pathway schematic png figures
xyz = ( 0.265, 0.47, 0.04 )
for j, strain in enumerate (['LM','LT']):
    for i, path in enumerate(['1','2','3','3R']):        
        pot_index=(j*2)+1
        ene_index=(j*2)
        image_filename="{}/{}_{:4.2f}.png".format(pathway_dir, path, conc)
        place_image( fig, axes[pot_index,i], image_filename, *xyz )
        place_image( fig, axes[ene_index,i], image_filename, *xyz )
fig.savefig( filename = '{}/neb_pathways_aniso.pdf'.format( figures_dir ) )    

set_rcParams( master_formatting )
````

````python
'''
   Produces Fig 5 in manuscript
'''
df = pd.read_csv( '{}/spinel_iso_pot_vol.csv'.format( data_dir ), delim_whitespace=True, comment='#')

fontsize = 18
scatter_formatting ={ 'font.size': fontsize,
                      'axes.labelsize': fontsize,
                      'xtick.labelsize': fontsize,
                      'ytick.labelsize': fontsize,
                      'lines.markersize': 10.0 }
set_rcParams( scatter_formatting )
      
fig, axes = plt.subplots(nrows=1, ncols=1, figsize = (5.5, 5.5))

x_dum=[]
y_dum=[]

x = np.array( [ 0.0, 1.0 ] )
y = np.array( [ 0.0, 1.0 ] )
axes.plot(x,y,'--',c='grey')

for conc in [ 0.25, 0.5 ]:
    for path in ['1','2','3','3R']:
        x=[]
        y=[]

        x2=[]
        y2=[]
        
        for strain in ['U','LM','LT']:
            loc = ( df.path == path ) & ( df.conc == conc ) & ( df.strain == strain )
            x.append( df[loc].dPolyV.values[0] ) 
            x_dum.append( df[loc].dPolyV.values[0] ) 
    
        U_max_ind = pd.DataFrame(data_spinel.loc[(data_spinel['path'] == path)
               &(data_spinel['conc']       == conc)
               &(data_spinel['strain']     == 'U')]['delta_E']).idxmax()    
        
        U_pot  = data_spinel['pot'].values[U_max_ind]
        U_Ebar = data_spinel['delta_E'].values[U_max_ind]           
    
        for j, strain in enumerate (['U','LM','LT']):
            max_ind = data_spinel.loc[(data_spinel['path'] == path)
                              &(data_spinel['conc']        == conc)
                              &(data_spinel['strain']      == strain)]['delta_E'].idxmax()
 
            pot   = data_spinel['pot'].values[max_ind]
            pot_t = data_spinel[ ( data_spinel['path'] == path ) & 
                                 ( data_spinel['conc'] == conc) & 
                                 ( data_spinel['strain'] == strain) & 
                                 ( data_spinel['image'] == 0) ]['pot']
            E_bar = data_spinel['delta_E'].values[max_ind]
            p_bar = data_spinel['d_pot'].values[max_ind]
            
            y.append(pot-U_pot[0])
            y_dum.append(pot-U_pot[0])
            y2.append(p_bar)
            x2.append(E_bar)
                    
            axes.plot(p_bar, E_bar ,symbol[conc],linestyle='None',c=colours[strain], clip_on=False)

axes.set_xlabel( r'$\Delta\Phi^\mathrm{max}_\mathrm{Li}$ [V]' )
axes.set_ylabel( r'$\Delta E_\mathrm{NEB}$ [eV]' )
 
popt, pcov = curve_fit(line, x_dum ,y_dum )

axes.xaxis.set_major_locator(ticker.MultipleLocator(0.2))
      
xfine = np.linspace(0.0,3.2,100)    

fig.tight_layout() 
fig.savefig( filename = '{}/pot_vs_delta_e.pdf'.format( figures_dir ) )

set_rcParams( master_formatting )
````

````python
'''
Produces pdfs for Figure 6
'''

set_rcParams( scatter_formatting )

fig, axes = plt.subplots(nrows=2, ncols=1, figsize = (5.5, 11))
title = [ '(a)', '(b)' ]

df = pd.read_csv( '{}/spinel_iso_pot_vol.csv'.format( data_dir, conc ), delim_whitespace=True, comment='#' )

for i, conc in enumerate( [ 0.25, 0.50 ] ):
    
    x=np.array( [ -0.2, 3.0 ] )
    y=np.array( [ 0.0, 0.0 ] )
    
    axes[i].plot(x,y,c='white', linewidth=5)

    axes[i].set_xlabel( '$\Delta V_\mathrm{oct} [\mathrm{\AA}^3]$' )
    axes[i].set_ylabel( r'$\Delta E_\mathrm{NEB}$ [eV]' )

    axes[i].set_xlim([0.0,3.0])
    axes[i].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

    for path in ['1', '2', '3','1R','2R','3R']:
        for ax in range(1,4):
            x=[]
            y=[]
            x.append(0.000)
            y.append(0.000)
        
            for strain in ['LM','LT']: 
                loc = ( df.path == path ) & ( df.conc == conc ) & ( df.strain == strain )
                x.append( df[ loc ].dPolyV.values[0] )
                y.append( df[ loc ].dEbar.values[0] )
            axes[i].plot( x, y, c='grey', linestyle='dashed' )
       
    x=[]
    y=[]
    x.append(0.000)
    y.append(0.000)

    for strain in (['LM','LT']):
        loc = ( df.conc == conc ) & ( df.strain == strain )
        x.append( df[ loc ].dPolyV.mean() )
        y.append( df[ loc ].dEbar.mean() )
 
    axes[i].plot(x,y,'co',c=nearly_black, linestyle='solid',label='average', clip_on=False)
    
    lm_lab = False
    lt_lab = False
    label = None
    
    for ax in range(1,4):
        for path in (['1','1R','2','2R','3','3R']):
            for j,strain in enumerate(['LM','LT']):
                col = colours[ strain ]
                if strain =='LT':
                    if lt_lab == True:
                        label = None
                    else:    
                        label = r"LiTi$_5$O$_{12}$"
                        lt_lab = True
                
                if strain == 'LM':
                    if lm_lab == True:
                        label = None
                    else:    
                        label = r"LiMn$_2$O$_4$"
                        lm_lab = True
                          
                x=[]
                y=[]
                loc = ( df.path == path ) & ( df.conc == conc ) & ( df.strain == strain )
                x.append( df[ loc ].dPolyV.values[0] )
                y.append( df[ loc ].dEbar.values[0] )       
     
                axes[i].plot(x,y,'ro', c=col, label=label, alpha=0.8)       

    axes[i].set_title( title[i], fontsize=fontsize, loc='left') 
    
axes[0].set_ylim([-0.30,0.15])
axes[1].set_ylim([-0.40,0.15])

plt.tight_layout()
plt.savefig( filename = '{}/oct_vol_vs_delta_e_iso.pdf'.format( figures_dir, conc ) )
    
set_rcParams( master_formatting )
````
